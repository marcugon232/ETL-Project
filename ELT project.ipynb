```python
import pandas as pd
```

```python
# Query All Records in the Zip Table
demo_data = pd.read_excel("Austin zip demo.xlsx")
rest_data = pd.read_csv("Austin_1000_Restaurants.csv")
```

```python
demo_data.head()
```

```python
demo_data['Key']=demo_data['Key'].astype(int)
```

```python
rest_data.head()
```

```python
rest_data['Zip Code'] = pd.to_numeric(rest_data['Zip Code'], errors='coerce')
```

```python
rest_data = rest_data.dropna(subset=['Zip Code'])
```

```python
# Merge the columns
combined_data = pd.merge(demo_data, rest_data, left_on="Key", right_on = "Zip Code", how="inner")
```

```python
# Combined Data
combined_data.head()
```

```python
combo2 = pd.DataFrame(combined_data)
```

```python
combo2
```

```python
# Module used to connect Python with MongoDb
import pymongo

# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the 'classDB' database in Mongo
db = client.ETLProject

collection = db.test
```

```python
collection.insert_many(combo2.to_dict('records'))
```

```python
# Iterate through each student in the collection
data = db.test.find()

for x in data:
    print(x)
```
